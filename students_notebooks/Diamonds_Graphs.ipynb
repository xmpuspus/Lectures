```python
import numpy as NP
import pandas as P
import seaborn as sns
import matplotlib.pyplot as MPLT

sns.set_palette("husl")
```

Code Comments: `Describe codes using comments`

```python
 data = P.read_csv('../data/diamonds_regression.csv')
```

```python
MPLT.figure(figsize=[20, 20])
MPLT.subplot(221)
MPLT.title('Diamond Prices According to Length')
sns.scatterplot(x='x',y='price', color='r', data=data)

MPLT.subplot(222)
MPLT.title('Diamond Prices According to Width')
sns.scatterplot(x='y',y='price', color='r', data=data)

MPLT.subplot(223)
MPLT.title('Diamond Prices According to Depth')
sns.scatterplot(x='z',y='price', color='r', data=data)

MPLT.subplot(224)
MPLT.title('Diamond Prices According to Carat')
sns.scatterplot(x='carat',y='price', color='r', data=data)
```

```python
MPLT.figure(figsize=(12, 12))
sns.pairplot(data[['x','y','z','carat','price']])
```

```python
MPLT.figure(figsize=[12, 12])
MPLT.subplot(221)
MPLT.hist(data['depth'],bins=50, color='orange')
MPLT.ylabel('Frequency')
MPLT.title('Frequency of Diamond Depth')

MPLT.subplot(222)
MPLT.hist(data['price'],bins=50, color='y')
MPLT.ylabel('Frequency')
MPLT.title(' Frequency of Diamond Prices')

MPLT.subplot(223)
MPLT.hist(data['carat'],bins=50, color='r')
MPLT.ylabel('Frequency')
MPLT.title('Frequency of Diamond Carat Weights')
```

```python
MPLT.figure(figsize=[12, 12])
MPLT.subplot(221)
MPLT.title('Diamond Prices According to Cut')
sns.barplot(x = 'cut', y = 'price', data=data, palette='inferno')
MPLT.subplot(222)
MPLT.title('Diamond Prices According to Cut')
sns.boxplot(x='cut', y='price', data=data, palette='inferno')
```

```python
MPLT.figure(figsize=[8, 8])
data['cut'].value_counts().plot(kind='bar' , color='r')
MPLT.title('Frequency of Diamond Cuts')
```

```python
MPLT.figure(figsize=[12, 12])
MPLT.subplot(221)
MPLT.title('Diamond Prices According to Color')
sns.barplot(x = 'color', y = 'price', data=data, palette='inferno')
MPLT.subplot(222)
MPLT.title('Diamond Prices According to Color')
sns.boxplot(x="color", y="price", data=data, palette='inferno')
```

```python
MPLT.figure(figsize=[8, 8])
data['color'].value_counts().plot(kind='bar' , color='r')
MPLT.title('Frequency of Diamond Colors')
```

```python
MPLT.figure(figsize=[12, 12])
MPLT.subplot(221)
MPLT.title('Diamond Prices According to Clarity')
sns.barplot(x = 'clarity', y = 'price', data=data, palette='inferno')
MPLT.subplot(222)
MPLT.title('Diamond Prices According to Clarity')
sns.boxplot(x="clarity", y="price", data=data, palette='inferno')
```

```python
MPLT.figure(figsize=[8, 8])
data['clarity'].value_counts().plot(kind='bar' , color='r')
MPLT.title('Frequency of Diamond Clarities')
```

```python
MPLT.figure(figsize=[20,20])
MPLT.subplot(221)
MPLT.title('Diamond Prices')
sns.scatterplot(x='carat',y='price',marker='x', hue='color', style='cut', size='clarity', palette='inferno', data=data)
MPLT.subplot(222)
MPLT.title('Diamond Prices According to Carat and Cut')
sns.scatterplot(x='carat',y='price', hue='cut', palette='inferno', data=data)
MPLT.subplot(223)
MPLT.title('Diamond Prices According to Carat and Color')
sns.scatterplot(x='carat',y='price', hue='color', palette='inferno', data=data)
MPLT.subplot(224)
MPLT.title('Diamond Prices According to Carat and Clarity')
sns.scatterplot(x='carat',y='price', hue='clarity', palette='inferno', data=data)
```

```python
MPLT.figure(figsize=(12, 12))
MPLT.title('Heatmap of Depth, Carat, Table, Price, and Dimensions')
sns.heatmap(data[['depth' , 'carat' , 'table' , 'price' , 'x' , 'y' , 'z']].corr(), annot = True)
```
